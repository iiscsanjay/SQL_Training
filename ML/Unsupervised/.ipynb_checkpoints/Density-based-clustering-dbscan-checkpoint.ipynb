```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
import scipy.cluster.hierarchy as sch
```

```python
#help(AgglomerativeClustering)
```

```python
sns.set()
```

```python
#help(make_blobs)
```

```python
X, y = make_blobs(500, 2, centers=4,random_state=0,cluster_std = 0.60)
plt.scatter(X[:,0],X[:,1],s=50)
```

```python
clf1 = DBSCAN(eps=0.724, min_samples=5)
clf1.fit(X)
labels = clf1.labels_
np.unique(labels)
```

```python
plt.scatter(X[labels==0, 0], X[labels==0, 1], s=50, marker='o', color='red')

plt.scatter(X[labels==0, 0], X[labels==0, 1], s=50, marker='o', color='red')
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=50, marker='o', color='blue')
plt.scatter(X[labels==2, 0], X[labels==2, 1], s=50, marker='o', color='green')
plt.scatter(X[labels==3, 0], X[labels==3, 1], s=50, marker='o', color='purple')
plt.show()
```

```python
clf2 = AgglomerativeClustering(n_clusters = 4)
clf2.fit(X)
label2 = clf2.labels_
np.unique(label2)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.scatter(X[labels==0, 0], X[labels==0, 1], s=50, marker='o', color='red')
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=50, marker='o', color='blue')
plt.scatter(X[labels==2, 0], X[labels==2, 1], s=50, marker='o', color='green')
plt.scatter(X[labels==3, 0], X[labels==3, 1], s=50, marker='o', color='purple')
plt.title('Clustering based on DBSCAN')
plt.subplot(1,2,2)
plt.scatter(X[label2==0, 0], X[label2==0, 1], s=50, marker='o', color='red')
plt.scatter(X[label2==1, 0], X[label2==1, 1], s=50, marker='o', color='blue')
plt.scatter(X[label2==2, 0], X[label2==2, 1], s=50, marker='o', color='green')
plt.scatter(X[label2==3, 0], X[label2==3, 1], s=50, marker='o', color='purple')
plt.title('Clustering based on Agglomerative')
plt.show()
```

```python
rng  = np.random.RandomState(1)
X = np.dot(rng.rand(10,2),rng.randn(2,200)).T
X.shape
```

```python
plt.scatter(X[:,0],X[:,1])
plt.axis('equal')
plt.show()
```

```python
pca = PCA(n_components=2)
pca.fit(X)
```

```python
pca.components_
```

```python
pca.explained_variance_
```

```python
#help(pca)
print(X.shape)
```

```python
X_pca = pca.transform(X)
X_pca.shape
```

```python
X_New = pca.inverse_transform(X_pca)
X_New.shape
```

```python
plt.scatter(X[:,0], X[:,1])
plt.scatter(X_New[:,0],X_New[:,1])
```

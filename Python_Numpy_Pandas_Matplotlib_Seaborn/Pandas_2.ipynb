```python
Exploratory Data Analysis : Outlier in the data 
Range : 
Q1 : 25% Quartile Range
Q2 : 50% Quartile Range
Q3 : 75% Quartile Range
Lowest Threshold for Outlier on left side : Q1 - 1.5 * IQR
Upper Threshold for Outlier on right side : Q3 + 1.5 * IQR
IQR =  Q3-Q1

0 ....25,,,,,,,,,,,,50,,,,,,,,75,,,,,,,100
```

```python
Q1 = df.describe()['Pregnancies'].loc['25%']
Q3 = df.describe()['Pregnancies'].loc['75%']
IQR = Q3-Q1
LT = Q1 - 1.5*IQR
UT = Q3 + 1.5*IQR
```

```python
UT
```

```python
np.sum(df.Pregnancies > UT)
```

```python
Q1 = df.describe()['BloodPressure'].loc['25%']
Q3 = df.describe()['BloodPressure'].loc['75%']
IQR = Q3-Q1
LT = Q1 - 1.5*IQR
UT = Q3 + 1.5*IQR
```

```python
np.sum(df.BloodPressure < LT)
```

```python
np.sum(df.BloodPressure > UT)
```

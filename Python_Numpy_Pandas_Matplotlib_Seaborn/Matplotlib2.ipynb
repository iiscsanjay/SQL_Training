```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
 
 
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']
speed = [23, 17, 35, 29, 12, 41]
 
 
# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)
 
# Creating color parameters
colors = ( "orange", "cyan", "brown", "grey", "indigo", "beige")
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
 
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:} g)".format(pct, absolute)

# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(speed, 
                                  #autopct='%1.1f%%',
                                  autopct=lambda pct: func(pct, speed),
                                  explode = explode, 
                                  labels = cars,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
 
# Adding legend
ax.legend(wedges, cars, title ="Cars", bbox_to_anchor =(1, 0, 0.5, 1))
 
#plt.setp(autotexts, size = 8, weight ="bold")
#ax.set_title("Customizing pie chart")
 
# show plot
plt.show()
```

```python
### 6. Box Plot : # Outlier Detection : q1, q3, IQR, LT, UT
df = pd.read_csv('diabetes.csv')
df.head()
plt.boxplot(df.Pregnancies)
plt.show()
```

```python
newdf = df.drop(['Outcome'],axis=1)
list(newdf.columns)
```

```python
# Because of null values was present in that series, it was not plotting the box plot
newdf.dropna(inplace=True)
```

```python
plt.figure(figsize = (15,15))
for i,col in enumerate(list(newdf.columns)):
    plt.subplot(4,2,i+1)
    plt.boxplot(x = col, data = newdf)
    plt.xlabel(f'{col}',fontweight='bold',fontsize=12)
plt.show()
```

```python
### Bar Stacked and Bar Multiplot together
processList = ['Process 1', 'Process 2', 'Process 3', 'Process 4', 'Process 5']
algoList = ['FCFS','SJF','Priority', 'RR','Priority RR' ]
```

```python
# Bar Grouped Plot
fcfscompletionTime = [15,64,69,39,30]
sjfcompletionTime = [15,69,20,29,44]
prioritycompletionTime = [60,45,20,69,15]
rrcompletionTime = [49,69,29,46,60]
priorityrrcompletionTime = [69,45,20,64,15]
myList = []
for i in range(5):
    myList.append([])
    myList[i].append(fcfscompletionTime[i])
    myList[i].append(sjfcompletionTime[i])
    myList[i].append(prioritycompletionTime[i])
    myList[i].append(rrcompletionTime[i])
    myList[i].append(priorityrrcompletionTime[i])
print(myList)
```

```python
# set width of bar 
barWidth = 0.15
fig, ax = plt.subplots(figsize =(12, 8)) 

# set height of bar 
Process1 = myList[0]
Process2 = myList[1]
Process3 = myList[2]
Process4 = myList[3]
Process5 = myList[4]
 
# Set position of bar on X axis 
br1 = np.arange(len(Process1))  # 0,1,2,3,4
br2 = [x + barWidth for x in br1]  #0.15,1.15,2.15,3.15,4.15
br3 = [x + barWidth for x in br2]  # 0.30, 1.30,
br4 = [x + barWidth for x in br3] 
br5 = [x + barWidth for x in br4] 

# Make the plot
plt.bar(br1, Process1, color ='r', width = barWidth, edgecolor ='grey', label ='Process1') 
plt.bar(br2, Process2, color ='g', width = barWidth, edgecolor ='grey', label ='Process2') 
plt.bar(br3, Process3, color ='b', width = barWidth, edgecolor ='grey', label ='Process3') 
plt.bar(br4, Process4, color ='k', width = barWidth, edgecolor ='grey', label ='Process4') 
plt.bar(br5, Process5, color ='c', width = barWidth, edgecolor ='grey', label ='Process5') 


# Adding Xticks 
plt.xlabel('Processes', fontweight ='bold', fontsize = 15) 
plt.ylabel('Time in Milliseconds', fontweight ='bold', fontsize = 15) 
#plt.xticks(enteries, processList,rotation=45, ha='right')

plt.xticks([r + barWidth for r in range(len(Process1))], 
        ['FCFS','SJF','PRIORITY', 'RR','PRIORITYRR'],rotation=45)
plt.title('Bar Plot for Scheduling in case 5 where Arrival Time [0,5,6,2,0] and Priority [1,2,3,1,2].')
plt.legend()
plt.savefig("case12.jpg", format="jpg") # specify filetype explicitly
plt.show() 
plt.close()
```

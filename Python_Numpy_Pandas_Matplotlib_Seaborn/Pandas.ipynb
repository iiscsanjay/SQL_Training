##### **Label**

##### **loc**

```python
# Multiple Rows Together Label Accessing
df.loc[['cobra','viper']]
```

```python
# Accessing a value with certain row and column
df.loc['viper','shield']
```

##### **iloc**

```python
df.iloc[0]
```

```python
df.iloc[1,1]
```

```python
df.loc['cobra']
```

```python
#Boolean way of accessing data frame worws
df.loc[[True,False,True,True]]
```

```python
# Writing a logic to get the rows which we want
l = []
for i in range(4):
    if i%2 == 1 :
        l.append(True)
    else:
        l.append(False)
        
df.loc[l]
```

```python
#Writing Condition statement to select Dataframe by loc function
df.loc[df['shield'] > 0]
```

```python
#Writing statement to select Dataframe by loc function cond1 | cond2 (or condition)
df.loc[(df['shield'] > 4 ) | (df['max_speed'] > 4)]
```

```python
#Writing statement to select Dataframe by loc function cond1 & cond2 (And condition)
df.loc[(df['shield'] > 4 ) & (df['max_speed'] > 4)]
```

```python
#Writing statement to select Dataframe by loc function cond1 ^ cond2 (Exor condition)
df.loc[(df['shield'] > 4 ) ^ (df['max_speed'] > 4)]
```

```python
# Numpy:
    - View and Copy
```

```python
r = [1, 2]
ser = pd.Series(r, copy=False)
print(f'Series Object:\n{ser}')
ser.iloc[0] = 999
print(f'{ser}')
print(f'{r}')
```

```python
r = [1, 2]
ser = pd.Series(r, copy=True)
print(f'Series Object:\n{ser}')
ser.iloc[0] = 999
print(f'{ser}')
print(f'{r}')
```

**When you are reating Panda Data Series from list, It doesn't modify the list variable even copy is True or False**

```python
r = np.array([1, 2])
ser = pd.Series(r, copy=False)
print(f'Series Object:\n{ser}')
ser.iloc[0] = 999
print(f'{ser}')
print(f'{r}')
```

```python
r = np.array([1, 2])
ser = pd.Series(r, copy=True)
print(f'Series Object:\n{ser}')
ser.iloc[0] = 999
print(f'{ser}')
print(f'{r}')
```

**When you are reating Panda Data Series from numpy array, It modify the numpy array in case when copy was false, else it doesn;t modify the value**

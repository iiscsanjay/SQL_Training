### Correct the Wrong data in large dataset (Outlier Correction) 
- Q1 : 25% Quartile Range
- Q3 : 75% Quartile Range
- IQR = Q3-Q1
- Lower Threshold : Q1 - 1.5 * IQR
- Upper Threshold : Q3 + 1.5 * IQR
---0-----25 -----50----75----100 ---

```python
# Will return all the header of the column names
df.columns
```

```python
# Removing Outlier in Sales data by putting np.NaN value and later removing those rows by calling 
# dropna function on dataframe
Q1 = df.describe()['Sales']['25%']
Q3 = df.describe()['Sales']['75%']
IQR = Q3 - Q1
LT = Q1 - 1.5*IQR
UT = Q3 + 1.5*IQR
```

```python
# First Method: Wrong Data : Replacing the Outlier values with Nan and removing it
df.Sales[df.Sales > UT] = np.NaN
df.dropna(inplace=True)
```

```python
# Second Method to replace Wrong Data :
#Iterating the DataFrame by Label and correcting the value
for index in df.index:
    if df.loc[index,'Sales'] > UT:
        df.loc[index,'Sales'] = UT
```

```python
# Third Method to Replace Wrong Data :
#Iterating the DataFrame by Label and removing the rows
for index in df.index:
    if df.loc[index,'Sales'] > UT:
        df.drop(index,inplace=True)
```

### MatplotLib important features :
- figure: you can change the figsize  which is width and height in inches and dpi
- show: will display the plot in gui window.
- savefig : save the plot into a file in local working directory
- title  : It will assign title given by string
- xlabel : It will assign x axis given by string
- ylabel : it will assign y axis given by string
- ylim : It will select the range for y axis for plotting (min,max)
- xlim : It will select the range for x axis for plotting (min,max)
- configure fonts in xlabel, ylabel, title
- xticks: It will configure the label values on the X Axis
- yticks: It will configure the label values on the Y Axis
- subplots
    - How many plots do you want to plot: 
    - How many Rows should be there and how many column columns? (1,4), (2,2), (4,1)
    - Each subplot should have a unique number (1,2,3,4)
- plot function:  Line, or Point Plot
    - marker
    - markersize
    - linestyles
    - linewidth
    - color
    - label
    - legend
        - ncol
        - loc
        - bbox_to_anchor
    - grid
    - plt.scatter(x,y) : plot individual points as marker ='.'
#### Plots in Matplotlib
- plt.figure() : plt functions
- fig, ax = plt.subplots() : ax functions : ax.set_xlabel, ax.set_title

```python
x = np.array([4,8])
y = np.array([11,22])

plt.figure(figsize=(8,6))
#plt.plot(x,y,linestyle=':',linewidth='4')
plt.plot(x,y,marker='s',markersize=10)
plt.xticks(ticks=np.arange(4,10,4.0),labels=["A","B"])
plt.yticks(np.arange(10,24,1.0),rotation=90)
#plt.plot(x,y,marker='s',markersize=15,color='y')
plt.title('Plot between X and Y for Line Plot',loc='center',fontdict =titleDict)
plt.xlabel('X Axis',fontdict=labelDict)
plt.ylabel('Y Axis',fontdict=labelDict)
plt.savefig('LinePlot.jpg')
```

### Labels

```python
x1 = np.array([4,8])
y1 = np.array([11,22])
y2 = np.array([ 15 for i in range(50)])
x2 = np.arange(4,8,0.2)
y3 = 2 * x2 + 3
#plt.plot(x,y,linestyle='-.',linewidth='12')
#plt.plot(x,y,marker='s',markersize=15)
plt.plot(x1,y1,color='b',label='Y=mx+c')
plt.plot(y2,color='r',label='parallel to x axis')
plt.plot(x2,y3,color='c',label='y2=mx+c')

plt.title('Plot between X and Y for Line Plot',loc='center',fontdict =titleDict)
plt.xlabel('X Axis',fontdict=labelDict)
plt.ylabel('Y Axis',fontdict=labelDict)
plt.legend(loc=(0.5,0.5))
plt.grid()
plt.tight_layout()
plt.show()
#plt.savefig('LinePlot.jpg')
```

```python
##### X values are generated between 5 to 51
x = np.arange(0, 100.0, 0.1)

# Y value are generated by formula for Sin function by computing radian from angle

y1 = 0.8 * x + 1
y2 = 0.2 * x + 0.5
y3 = 5 * x + 12
y4 = 2.5 * x + 20

# any connnected plot between x and y function in 2 d space
plt.plot(x,y1,color='r', label='y=0.8*x+1')
plt.plot(x,y2,color='g', label='y=0.2*x+0.5')
plt.plot(x,y3,color='b', label='y=5*x+12')
plt.plot(x,y4,color='c', label='y=2.5*x+20')
plt.legend(loc='best', ncol=1)
#plt.legend(bbox_to_anchor=(0,1), ncol=4)
plt.title('Multiple function plotting')
# Limiting your plot space by x-range and y-range
plt.xlim((20,60))
plt.ylim((0,200))
plt.xlabel('X')
plt.ylabel('Y')
#plt.tight_layout()
plt.show()
```

```python
# X values are generated between 5 to 51
x = np.arange(0, 30.0, 0.1)

y1 = 0.8*x + 1
y2 = np.sin(x)
y3 = np.cos(x)
y4 = np.square(x) + x


plt.figure(figsize=(10,4))
plt.subplot(1,4,1)
plt.plot(x,y1,color='r', label='y=0.8x+1')
plt.xlabel('X')
plt.ylabel('Y=0.8x+1')
plt.title('Linear Plot')

plt.subplot(1,4,2)
plt.plot(x,y2,color='g', label='y=sinx')
plt.xlabel('X')
plt.ylabel('Y=sinx')
plt.title('Sin Wave')

plt.subplot(1,4,3)
plt.plot(x,y3,color='b', label='y=cox')
plt.xlabel('X')
plt.ylabel('Y=cosx')
plt.title('Cos Wave')

plt.subplot(1,4,4)
plt.plot(x,y4,color='c', label='y=x^2+x')
plt.xlabel('X')
plt.ylabel('Y=X^2')
plt.title('Parabola Function')

#plt.suptitle('Plotting 4 different mathematical functions',fontdict=titleDict)
#plt.subplots_adjust(left=0.1,bottom=0.1,right=1.0, top=0.5, wspace=0.4)
#plt.tight_layout()
```

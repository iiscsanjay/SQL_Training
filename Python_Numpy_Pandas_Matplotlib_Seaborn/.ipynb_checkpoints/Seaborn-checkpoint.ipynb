### Seaborn Function Module:
1. Different Color Combinations for Plot: Palette
    - color_palette() : Default Color Map of 10 colors with their (RGB) value as tuple in the list
    - set_palette('palette_name') : It will apply the palette_name of selected color theme to current plot
    - palplot(sequence_colors) : plot the values in a color palette as a horizontal array.
    - set_style() : Set the parameters that control the general style of the plots (darkgrid, whitegrid, dark, white, ticks)
2. Provides Different Datasets which we can load at any time :
    - get_dataset_names(): returns all the dataset names presented in the seaborn module
    - load_dataset(): able to load the dataset with the given name
3. Different Plots:
    - Histogram Plot: histplot(): 
    - Bar Plot: barplot():
    - Displot: displot() : It can be used to call hist,kde, ecdf plot
    - Line Plot: lineplot()
    - Box plot: boxplot() :
    - Violin Plot: violinplot(): 
    - Scatter Plot : scatterplot()
    - Joint Plot : jointplot()
    - Pair Plot: pairplot()
    - Strip Plot: stripplot()
    - Swarm Plot : swarmplot()
    - HeapMap Plot: heatmap()
    - Linear Relationship plot : lmplot()
    - Non-Linear Relation Plot : regplot()

### Seaborn Color Palettes Choices:
- **Qualitative Color** :
  *  **Default Categorical** : deep, muted, bright, pastel, dark, colorblind, tab10, tab20, tab20c, tab20b
  *  **Circular palettes** : hls, husl
  *  **Brower palettes** : Set1, Set2, Set3, Paird, Accent, Pastel1, Pastel2, Dark2 
- **Sequential Color** :
   * **Default**  : rocket, mako, flare,crest, viridis, plasma, inferno, magma, cividis.

- **Divering Color** :
   * **Default**  : vlag, icefire, coolwarm, bwr, seismic
   *  **Brower palettes** : RdBu, RdGy, PRGn, PiYG, BrBG, RdYlBu, RdYlGn, Spectral

```python
def sin_plot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,8):
        plt.plot(x, np.sin(x+i*0.5) * (8-i)* flip, label=f'y{i}')
    plt.legend(loc=(1,0.1))
sns.set_style('whitegrid')
sns.set_palette('hls',n_colors=14)
sin_plot()
```

```python
def sin_plot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,15):
        plt.plot(x, np.sin(x+i*0.5) * (15-i)* flip, label=f'y{i}')
    plt.legend(loc=(1,0.1))
sns.set_style('white')
color = sns.color_palette(palette='hls',n_colors=14)
sns.set_palette(color)
sin_plot()
```

### 12 : Heatmap plot : Correlation Matrix values in the plot

```python
X : Random Variable: Mean, Standard Derivation, Variance ()
X,Y : Random Variable: Co-Variance () : constant
Correlation Matrix  :(X,Y,Z,W) : 4x4 : Co-variance (X,X) = 1.0  

Co-Variance Value > 0.6 < -0.6  is considered in statistics as high covariant variables.

When two variables are highly correlated, we say there is collinearity.  This kind of behaviour is known
as multi-collinearity.

Machine Learning : solve 2 kinds of problem:
1. Regression Problem:  Sales Prediciton, Weather temp. (Numerical value prediction )
2. Classification Problem: Human body parameters (Glucose, bmi, insulin, ): Diabetics(1)/Non-Diabtics (0)
```

```python
penguins.describe().corr()

# Wallmart: Sale prediction (Dependent Variable) :  (Independent Variable ) Day, Holiday, Store location, Number of staff
#  Dependentvariable = M1 * IV1 +  M2 *IV2
#Y (body_mass_g) =  M1 * bill_length_mm +  M2 * flipper_length_mm + C
# Y (body_mass_g) =  M1 * bill_depth_mm  +  M2 * flipper_length_mm + C
```

```python
# Systematic sampling
samples = []
while len(samples) != 100:
    value = np.random.randint(0,768,1)
    if value[0] % 2  == 0:
        samples.append(value[0])
print(samples)
```

```python
df2 = df.iloc[samples]
df2.Outcome.value_counts()
```

```python
import random
dlist = list()
while len(dlist) != df2.Outcome.value_counts()[0] :
    index = random.choice(samples)
    if df.iloc[index].Outcome == 1:
        dlist.append(index)

for index in samples:
    if df.iloc[index].Outcome == 0:
        dlist.append(index)
```

### Probability Distribution:
    - Continuous  Distribution: Normal Distribution
    - Discrete Distribution : Poisson and Binomial Distribution

```python
##### Normal Distribution
mu, sigma = 0, 0.1 # mean and standard deviation
y = np.random.normal(mu, sigma, 1000)
sns.displot(x=y, kind='hist',kde=True)
```

```python
#### Bionomial Distribution
n, p = 10, .5  # number of trials, probability of each trial
y = np.random.binomial(n, p, 1000)
sns.displot(x=y, kind='hist',kde=True)
```

```python
# importing poisson from scipy
from scipy.stats import poisson
 
# importing numpy as np
import numpy as np
 
# importing matplotlib as plt
import matplotlib.pyplot as plt
 
 
# creating a numpy array for x-axis
x = np.arange(0, 20, 1)
 
# poisson distribution data for y-axis
y1 = poisson.pmf(x, mu=1)
y2 = poisson.pmf(x, mu=4)
y3 = poisson.pmf(x, mu=10)

 
# plotting the graph
plt.plot(x, y1, 'rd-',label='Lambda = 1')
plt.plot(x, y2, 'gd:',label='Lambda = 4')
plt.plot(x, y3, 'kd--', label='Lambda = 10')
plt.legend()

# showing the graph
plt.show()
```
